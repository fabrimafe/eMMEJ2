# Opening the MMEJ detection algorithm output data
## This notebook is an example of how to properly open the MMEJ detection algorithm output data

```python
# importing relevant libreries
import pandas as pd
import numpy as np
import io

# set pandas display options
pd.options.display.max_colwidth = 2200
pd.set_option("display.max_columns", None)
```

```python
# Set path to output file
path_to_df = r'/home/labs/alevy/amitcu/MMEJ_detection_algorithm/CRIL_sample/CRIL_13-Int_MMEJdf.csv'
# set columns Dtypes
dtypes = { 'Unnamed: 0' : 'int64', 'Chr' : 'str', 'Start' : 'int64', 'end' : 'int64' ,
          'Ref' : 'str', 'Alt' : 'str', 'ancesstral_allele' : 'str', 'derived_allele' : 'str',
          'derived_allele_freq' : 'float64', 'indel_type' : 'category', 'indel_pos' : 'int64',
          'indel_length' : 'int64', 'indel_len_bin' : 'category', 'mmej': 'category', 'fasta_context_position' : 'str',
          'fasta_context_position_2Kbp' : 'str', 'accession_context' : 'str', 'true_rep_micro_rep_INS' : 'str', 
          'mmej_cand' : 'str', 'S_mmej_cand' : 'str', 'S_mmej_cand_len' : 'int64', 'S_mmej_cand_len' : 'float64',
          'mmej_cand_len' : 'float64', 'tamplate_switch_repeat' : 'str', 'mmej_marked' : 'str', 'mmej_marked_on_ref' : 'str',
          'last_dimer' : 'str', 'SD_mmej_trans' : 'bool', 'SD_mmej_snap_back' : 'bool', 'S_tamplate_switch_repeat' : 'str',
          'S_true_rep_micro_rep' : 'str', 'S_last_dimer' : 'str', 'S_microhomology_marked' : 'str', 'fasta_context_seq_120' : 'str',
          'fasta_context_seq_2K' : 'str', 'mmej_probability' : 'float64', 'SD_mmej_snap_back_prob' : 'float64' }

# loading the data
df = pd.read_csv(path_to_df, sep = '\t', dtype = dtypes)
print(df.info())

# Since the 'mmej' columns isn't a real boolian, (it contains also the 'potential_INS_MMEJ'),
# for now, we can reat them as 'False'.
df['mmej'].replace(to_replace='True', 
                            value=True, inplace = True)
df['mmej'].replace(to_replace='False', 
                            value=False, inplace = True)
df['mmej'].replace(to_replace='potential_INS_MMEJ', 
                            value=False, inplace = True)
df['mmej'] = df['mmej'].astype('bool')


df['mmej_probability'] = 1 - df['mmej_probability']
df['SD_mmej_snap_back_prob'] = 1 - df['SD_mmej_snap_back_prob']
df.loc[:,'mmej_probability'].fillna(value = 0, inplace=True)
```

```python
df.loc[((df['mmej'] == True) & (df['indel_type'] == 'DEL')), ['REF', 'ALT','Read','reference_context_seq_120', 'mmej_marked_on_ref', 'mmej_marked', 'mmej_probability']].sample()
```
